```python
import os
import numpy as np
import pandas as pd
```

```python
df_asym = pd.read_csv('../symmetry/symmetry_mean_ratings.csv')
df_asym['diff'] = df_asym.forward_rating - df_asym.backward_rating
df_asym['diff'] = df_asym['diff'].abs()
print(df_asym.shape)
df_asym.head()
```

```python
df_asym_top = df_asym[df_asym['diff'] > 0.1].copy()
print(df_asym_top.shape)
df_asym_top
```

```python
df_raw = pd.read_csv('relsim_raw_data.csv')
print(df_raw[(df_raw.pair_left_word1=='dull')&(df_raw.pair_left_word2=='bright')].sim_rating.mean())
print(df_raw[(df_raw.pair_right_word1=='dull')&(df_raw.pair_right_word2=='bright')].sim_rating.mean())
```

```python
# lw1, lw2 = 'hairdresser', 'comb'
# rw1, rw2 = 'pitcher', 'baseball'
lw1, lw2 = 'mother', 'children'
rw1, rw2 = 'words', 'letters'

oneway = (df_raw.pair_left_word1==lw1)&(df_raw.pair_left_word2==lw2)&(df_raw.pair_right_word1==rw1)&(df_raw.pair_right_word2==rw2)
another = (df_raw.pair_right_word1==rw1)&(df_raw.pair_right_word2==rw2)&(df_raw.pair_left_word1==lw1)&(df_raw.pair_left_word2==lw2)

print(df_raw[oneway].sim_rating)
print(df_raw[oneway].sim_rating.mean())
print(df_raw[another].sim_rating.mean())
```

```python
for row in df_asym_top.iterrows():
    lw1 = row[1]['pair1_word1']
    lw2 = row[1]['pair1_word2']
    rw1 = row[1]['pair2_word1']
    rw2 = row[1]['pair2_word2']
    asym_diff = row[1]['diff']
    
    mask1 = ((df_raw.pair_left_word1==lw1) & 
             (df_raw.pair_left_word2==lw2) & 
             (df_raw.pair_right_word1==rw1) & 
             (df_raw.pair_right_word2==rw2))
    
    mask2 = ((df_raw.pair_left_word1==rw1) & 
             (df_raw.pair_left_word2==rw2) & 
             (df_raw.pair_right_word1==lw1) & 
             (df_raw.pair_right_word2==lw2))
    
    score1 = df_raw[mask1].sim_rating.mean()
    score1_n = df_raw[mask1].sim_rating.values.size
    score2 = df_raw[mask2].sim_rating.mean()
    score2_n = df_raw[mask2].sim_rating.values.size
    diff = score1-score2
    if not np.isnan(np.abs(diff)): 
        print(lw1, lw2, '     ', rw1, rw2)
        print(asym_diff, diff)#, score1_n, score2_n)
```
